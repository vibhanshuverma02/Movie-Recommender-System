# Movie Recommender System

```python
# Movie Recommender System

# Recommender Systems, also labelled as recommendation systems, are statistical algorithms the
# recommend rpoducts to users based on similarities between the buying trends of various user
# or similarities between the products

# Item-based collaborative filtering
```

```python
# Collaborative filtering- The process used to calculate similarities between the buying trends of various user
# or similarities between products is called collaboratiev filtering

# Classified into two types- User based collaborative filtering, Item based collaborative filtering
```

```python
# User based collaborative filtering- is dependent on user choices. For example, in a recommender system 
# based on user-based collaborative filtering, if two user X and Y, like products A and B
# and there is another user Z who likes product A, then the product B will be recommended to user Z
```

```python
# Item based collaborative filtering products are recommended based on similarities between themselves.
# For ex if a user likes product A and product A has properties X and Y, another product B with properties
# X and Y will also be recommended ot the user
```

# Item Based Collaborative Filtering

```python
# In item based colloborative filtering, products are recommended based on common characteristics.
# For, instance in a movie recommender system based on item-based collaborative filtering,
# a commong characteristic could be the director of a movie, the actors in the movie, and so on.

# In our datasetwe have average ratings for different movies given by different users. We will use the
# average ratings as the common characteristic of the collaborative filtering of movies

# The first step is to create a dataframe where each movie is represented by a column and rows contain
# user ratings for movies

user_movie_ratings_matrix= merged_movie_df.pivot_table(index="userId", columns="title", values="rating")

user_movie_ratings_matrix
```

```python
user_movie_ratings_matrix.shape

# Dataset contains 610 unique user and 9,719 unique movies
```

```python
# Now we will find movie recommendation based on a single movie and then based on multiple movies

# Finding recommendation based on a single movie- Suppose we want to find recommendation based on
# the movie Pulp Fiction (1994).

# First we will filter the column that contains the user ratings for the movie

pulp_fiction_ratings= user_movie_ratings_matrix["Pulp Fiction (1994)"]
```

```python
# Next, we will find the correlation betwen the user ratings of all the movies and the user ratings
# for the Pulp Fiction

pulp_fiction_correlation= pd.DataFrame(user_movie_ratings_matrix.corrwith(pulp_fiction_ratings), columns=["pf_corr"])
```

```python
pulp_fiction_correlation.sort_values("pf_corr", ascending=False).head(5)
```

```python
# Correlation itself is not giving meaningful results, one solution to this problem can be that
# in addition to the correlation between the movies, we also use rating counts, for the
# correlated movie as a criterion for finding the best recommendation

pulp_fiction_correlations= pulp_fiction_correlation.join(movie_rating_mean_count["rating_count"])

pulp_fiction_correlations.head()
```

```python
# Thepf_corr column contains some NaN values. This is because there can be movies that are rated
# by users who did not rate Pulp Fiction (1994). In such cases, correlation will be null.

# We will remove all the movies with null correlation with Pulp Fiction (1994)

pulp_fiction_correlations.dropna(inplace=True)
```

```python
pulp_fiction_correlations.sort_values("pf_corr", ascending=False).head(5)
```

```python
# A better way is to find the movies with rating counts of atleast 50
# and having the highest correlation with Pulp Fiction (1994)

pulp_fiction_correlations_50= pulp_fiction_correlations[pulp_fiction_correlations['rating_count']>50]

pulp_fiction_correlations_50.sort_values('pf_corr', ascending=False).head()
```

```python
# Finding recommendation based on multiple movies- In this we will see how to recommend movies to a user based on
# his ratings of multiple movies. The first step is to create a dataframe, which contains
# a correlation between all the movies in our dataset in the form of a matrix

all_movie_correlations= user_movie_ratings_matrix.corr(method='pearson', min_periods=50)

# The min_periods attrbute vaue specifies the minumum number of observations
# required per pair of columns to have a valid result. A min_periods value of 50 specifies
# calculating correlation for only those pairs of movies that have been rated by atleast 50 users.
# For, the rest of the movie pairs, the correlation will be null

all_movie_correlations.head()
```

```python
# Now suppose a new user logs into your website. The user has already watched three movies
# and has given rating to those movies

movie_data=[['Forrest Gump (1994)',4.0],['Fight Club (1999)',3.5],['Interstellar (2014)',4.0]]

test_movies= pd.DataFrame(movie_data, columns= ['Movie_Name','Movie_Rating'])

test_movies.head()
```

```python
# We will be recommending movies from our dataset based on the ratings by a new
# user for these three movies

print(test_movies['Movie_Name'][0])

print(test_movies['Movie_Rating'][0])
```

```python
# From all_movie_correlations dataframe, let's obtain correlation values for the movies related to
# Forest Gump (1994)

all_movie_correlations['Forrest Gump (1994)'].dropna()

```

```python
# Now, you know how to obtain names and ratings of movies from the test_movie dataframe
# and how to obtain correlations of all the movies with a single movie using the movie title

# Next, we will iterate through the three movies in the test_movie dataframe, find the correlated movies,
# and then multiply the correlation of all the correlated movies with
# the ratings of the input movie

# The correlated movies, alongwith the weighted correlation (calculated by multiplying the actual correlation
# with the ratings of the movies in the test_movies dataframe) are appended to an empty series
# names recommended movie

recommended_movies= pd.Series()

for i in range(0,2):
    movie=all_movie_correlations[test_movies['Movie_Name'][i]].dropna()
    movie= movie.map(lambda movie_corr: movie_corr*test_movies["Movie_Rating"][i])
    recommended_movies= recommended_movies.append(movie)
```

```python
recommended_movies
```

```python
# To get a final recommendation you can sort the movies in the descending order of the wighted correlaiton

recommended_movies.sort_values(inplace=True, ascending=False)

print(recommended_movies.head(10))
```

```python
def remove_punct(text):
    text_nopunct="".join([char for char in text if char not in string.punctuation])
    return text_nopunct

dataset['body_text_clean']=dataset['body_text'].apply(lambda x:remove_punct(x))

dataset.head()
```
